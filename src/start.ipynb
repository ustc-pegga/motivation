```python
def get_result(envpath,execute,datasets,device,threads):
    print('adb shell "{} && {} {} {} {}"'.format(envpath,execute,datasets,device,threads))
    result = os.popen('adb shell "{} && {} {} {} {}"'.format(envpath,execute,datasets,device,threads))
    context = result.read()
    return context
```

```python
# test5
datasets = "imagenet"
type = 'mbv2'
device = 'mate30e_core7'
src_path = "../data/test5/test5_{}/{}/ms".format(type,datasets)
dst_path = "/data/local/tmp/data/test5/test5_{}/{}".format(type,datasets)
txt_path = "../data/test5/test5_{}/{}/data/{}.txt".format(type,datasets,device)
def get_device(device):
    if "gpu" in device:
        return "1"
    else:
        return "0"
    
def make_dirs(path):
    if not os.path.exists(path):
        os.makedirs(path)

def start_test5(src_path,dst_path,txt_path,device):

    os.system("adb push {} {}".format(src_path,dst_path))

    result = get_result(envpath,execute,dst_path,get_device(device),threads)
    f = open(txt_path,'w')
    f.write(result)
    return result

result = start_test5(src_path,dst_path,txt_path,device)
print(result)
```
